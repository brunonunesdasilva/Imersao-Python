```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/acoes-alura.xlsx", sheet_name= "Principal")
df_principal.head(10)
```

```python
df_Totaldeacoes = pd.read_excel("/content/acoes-alura.xlsx", sheet_name= "Totaldeaçoes")
df_Totaldeacoes
```

```python
df_Ticker = pd.read_excel("/content/acoes-alura.xlsx", sheet_name= "Ticker")
df_Ticker
```

```python
df_Chat_GPT = pd.read_excel("/content/acoes-alura.xlsx", sheet_name= "Chat GPT")
df_Chat_GPT
```

```python
df_principal = df_principal[['Ativo', 'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns= {'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_Totaldeacoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtdeTeorica'}).copy()
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['qtdeTeorica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['qtdeTeorica'] = df_principal['qtdeTeorica'].astype(int)
```

```python
df_principal['Resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_Ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_Chat_GPT, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade em anos':'idade'}).copy()
```

```python
df_principal['Cat_idade'] = pd.cut(df_principal['idade'], bins=[-float('inf'), 49, 100, float('inf')], labels=['Menos de 50', 'Entre 50 e 100', 'Mais de 100 anos'])
df_principal
```

```python
# Calculando o maior valor
maior = df_principal['variacao_rs'].max()

# Calculando o menor valor
menor = df_principal['variacao_rs'].min()

# Calculando a média
media = df_principal['variacao_rs'].mean()

# Calculando a média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['variacao_rs'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['variacao_rs'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='variacao_rs', text='variacao_rs', title='Variacao Reais por resultado')
fig.show()
```
